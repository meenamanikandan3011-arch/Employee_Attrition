```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Employee-Attrition - Employee-Attrition.csv")
df
```

Basic Data Understanding

```python
# Shape of dataset
print("Rows, Columns:", df.shape)

# Summary of dataset
df.info()

# Check missing values
df.isnull().sum()
```

Handling Missing Values

```python
# Numeric columns - fill missing with median
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Categorical columns - fill missing with mode
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

```python
columns_to_drop = ['EmployeeCount', 'Over18', 'StandardHours']
df = df.drop(columns=columns_to_drop)
```

```python
#COLUMN NAME STANDARDIZATION
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

print("\nStandardized Columns:")
print(df.columns.tolist())
```

```python
# Install scikit-learn if missing and encode categorical variables
from sklearn.preprocessing import LabelEncoder 

# create a new LabelEncoder for each column to avoid leaking categories between columns
for col in df.select_dtypes(include='object'):
    le = LabelEncoder() # create a new LabelEncoder for each column
    df[col] = le.fit_transform(df[col]) # Encode categorical variables into numeric 
df
```

```python
df.to_csv("Cleaned_Employee_Attrition.csv", index=False)
print("Preprocessed dataset saved!")
```

```python
import numpy as np

# Select numerical columns except target variable
num_cols = [col for col in df.select_dtypes(include=['float64', 'int64']).columns if col != 'attrition']

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

print("Outlier treatment completed")
print(df['attrition'].value_counts())   # Should show both classes 0 & 1
```
