EDA helps understand employee behavior patterns and reveal the most influential factors driving attrition (employees leaving the company). By analyzing the dataset graphically and statistically, we find why employees leave and who is most at risk.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Cleaned_Employee_Attrition.csv")
```

```python
#1. Attrition Count
# column name in dataframe is 'attrition' (lowercase)
sns.countplot(data=df, x="attrition")
plt.title("Attrition Distribution (0=Stay, 1=Leave)")
plt.show()
```

```python
# 2. Salary vs Attrition
sns.boxplot(x="attrition", y="monthlyincome", data=df)
plt.title("Monthly Income vs Attrition")
plt.show()
```

```python
# 3. Job Satisfaction vs Attrition
sns.countplot(data=df, x="jobsatisfaction", hue="attrition")
plt.title("Attrition by Job Satisfaction Level")
plt.show()
```

```python
# 4. Department vs Attrition
sns.countplot(data=df, x="department", hue="attrition")
plt.xticks(rotation=45)
plt.title("Attrition by Department")
plt.show()
```

```python
# 5. Overtime vs Attrition
sns.countplot(data=df, x="overtime", hue="attrition")
plt.title("Attrition by OverTime")
plt.show()
```

```python
# 6. Correlation Heatmap
# Select only relevant variables for correlation
important_features = [
    "attrition",
    "monthlyincome",
    "jobsatisfaction",
    "environmentsatisfaction",
    "relationshipsatisfaction",
    "worklifebalance",
    "yearsatcompany",
    "yearssincelastpromotion",
    "totalworkingyears",
    "overtime",
    "age",
    "joblevel",
]

# Ensure only existing columns are used
important_features = [col for col in important_features if col in df.columns]

subset_df = df[important_features]

plt.figure(figsize=(10,6))
sns.heatmap(subset_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Important Features Only)")
plt.show()
```

The refined heatmap showed that low monthly income, low job satisfaction, frequent overtime, and long delay since last promotion are strongly correlated with attrition."
